# Depth First Search 

## EXAMPLE 1.

```python
# graph representation 
adj_list={
      "S": ["A","B"],
      "A": ["B","D"],
      "B": ["C"],
      "C": ["E"],
      "D": ["G"],
      "E": [],
      "G": []
    }
```

```python
# Vertices visited - empty set
visited = set()
```

```python
def dfs(visited,adj_list,root):
    if root not in visited:
        print(root)
        visited.add(root)
        for neighbour in adj_list[root]:
            dfs(visited,adj_list,neighbour)
dfs(visited,adj_list,"S")
```
